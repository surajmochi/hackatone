```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_data(soup3):
    cards = soup3.find_all("div",attrs={"class":"_1UoZlX"})
    for card in cards:
        title = card.find("div",attrs={"class":"_3wU53n"})
        price = card.find("div",attrs={"class":"_1vC4OE _2rQ-NK"})
        rating = card.find("div",attrs={"class":"hGSR34"})
        if rating:
            rating_text = rating_text
        else:
            rating_text = None
        print("{} {} {}".format(title.text,price.text,rating_text))

start_url = "https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
home_url = "https://www.flipkart.com"

count = 0

while count < 10:
    print("******************{}************".format(count)+"\n")
    response3 = requests.get(start_url)
    soup3 = BeautifulSoup(response3.content,"html.parser")
    get_data(soup3)
    next_link = soup3.find("a",text="Next")
    start_url = home_url+next_link.get("href")
    
    count +=1
```
