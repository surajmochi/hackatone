```python
class Course:
    courses = []
    def __init__(self,id,title,duration,author,price):
        self.c_id = id
        self.c_title = title
        self.c_duration = duration
        self.author = author
        self.price = price
        Course.courses.append(self)
    
#     def __str__(self):
#         return ;

class Tutorial:
    
    tutorial = []
    
    def __init__(self,id,title,content,course):
        self.t_id = id
        self.t_title = title
        self.t_content = content
        self.course = course
        Tutorial.tutorial.append(self)
     
    def get_details(self):
        return "{} {} {}".format(self.t_id,self.t_title,self.t_content)
    


course1 = Course(1,"python","30","ABC",5000)
course2 = Course(2,"python django","40","PQR",6000)
course3 = Course(1,"Networking","50","XYZ",7000)

tut1 = Tutorial(101,"python Datatype","list,Dict,set,tuple",course1)
tut2 = Tutorial(102,"python function","parameter passing",course1)
tut3 = Tutorial(103,"python django","Models and ORM",course2)
tut4 = Tutorial(104,"OSI model","7 Layers",course3)
tut5 = Tutorial(105,"static routing","floating routing",course3)

for course in Course.courses:
    if course.c_title == "python":
        c_python = course

print("Tutorial for course Python")        
for tut in Tutorial.tutorial:
    if tut.course == c_python:
        print(tut.get_details())
        
        
print(c_python.c_title)
#print(Tutorial.tutorial)

print("All tutorials for author ABC")
for tut in Tutorial.tutorial:
    if tut.course.author == "PQR":
        print(tut.get_details)
        

c_abc  = [course for course in Course.courses if course.author == "ABC"]
print(c_abc)
```
